```python
def cut_into_clips(directory, path:str, t_of_clip, t_start, t_start_2):
    #i = [x for x in range(len(os.listdir(directory)))]
    #i_filename = 0
    os.makedirs(path)
    #os.makedirs(f'Projekt_alternatywny/{name}')

    for file in os.listdir(directory):
        filename = os.fsdecode(file)
        #idx = i[i_filename]
        if filename.endswith('.mp4'):
            for t in range(10):
                # part 1
                if t < 5:
                    clip_start = t_start + t * t_of_clip
                    clip_end = clip_start + t_of_clip
                    ffmpeg_extract_subclip(directory + f"/{filename}", clip_start, clip_end , targetname=path + f"/{filename[:-3]}_{t}.mp4")
                # part 2
                if t >= 5:
                    clip_start = t_start_2 + ((t % 5) * t_of_clip)
                    clip_end = clip_start + t_of_clip
                    ffmpeg_extract_subclip(directory + f"/{filename}", clip_start, clip_end, targetname=path + f'/{filename[:-3]}_{t}.mp4')
```

```python
def check_time(directory, file):
    k = 1000000
    filename = os.fsdecode(file)

    print(f'Sprawdzenie czasu pliku {filename}')

    if filename.endswith('.aedat4'):
        with dv.AedatFile(directory + '/' + filename) as f:
            events = np.hstack([packet for packet in f['events'].numpy()])
            timestamps, x, y, polarities = events['timestamp'], events['x'], events['y'], events['polarity']

    file_time_start = timestamps[0]
    end_time = timestamps[-1]

    file_duration = end_time - file_time_start

    print('Ilość eventów', len(timestamps))

    print(f'Czas trwania:{file_duration // k} s')
```

```python
def cut_from_timestamps(clip_start, clip_end, timestamps, idx_start = 0, idx_end = 0):
    """

    :param clip_start: start klipu (mikrosekundy)
    :param clip_end:  koniec klipi (mikrosekundy)
    :param timestamps: np.array zawierający timestampy z pliku
    :param idx_start: początkowy index poszukiwania
    :param idx_end: końcowy index poszukiwania (niepotrzebny)
    :return: idx_start: indeks początkowy klipu , idx_end : indeks końcowy klipu
    """
    #idx_start, idx_end = 0, 0
    time = timestamps[0]

    end = len(timestamps)

    while time <= clip_start :
        idx_start +=1
        if idx_start == end - 1:
            print('Klip wykracza poza długość filmu')
            break
        time = timestamps[idx_start]
        #idx_start +=1

    print('start', idx_start)

    time = timestamps[idx_start]

    idx_end = idx_start

    while time <= clip_end:
        idx_end +=1
        if idx_end == end-1:
            print('Przekroczono długość filmu. Skrócono ostatni klip.')
            break
        time = timestamps[idx_end]
        #idx_end +=1

    print('end', idx_end)

    return idx_start, idx_end
```

```python
def cut_into_clips_kz(directory:str, destination:str, file, duration, start, start_2):
    """
    :param directory: źródło z którego pochodzi plik
    :param destination: miejsce zapisywania
    :param file: plik
    :param duration: czas trwania klipu (sekundy)
    :param start: start pierwszej części klipów (sekundy)
    :param start_2: start drugiej części klipów (sekundy)

    Funkcja tworzy pliki npy zawierające klipu z kamery zderzeniowej
    """

    k = 1000000 # mikrosekund w sekundzie, do konwersji

    filename = os.fsdecode(file)
    events = 0
    timestamps = 0

    print(f'Cutting clip from {filename}')

    if filename.endswith('.aedat4'):
        with dv.AedatFile(directory + '/' + filename) as f:
            events = np.hstack([packet for packet in f['events'].numpy()])
            timestamps, x, y, polarities = events['timestamp'], events['x'], events['y'], events['polarity']

    file_time_start = timestamps[0]
    end_time = timestamps[-1]

    file_duration = end_time - file_time_start

    print('Ilość eventów', len(timestamps))

    print(f'Czas trwania:{file_duration // k} s')

    start_time = file_time_start + start * k # klipu zaczynają się po tylu mikrosekundach
    start_2_time = file_time_start + start_2 * k # druga część klipów po tylu mikrosekundach
    duration = duration * k # trwa tyle mikrosekund

    idx_start, idx_end = 0, 0

    # Przypadki

    # Filmik nagrany pojedynczno, nie w serii
    if 0 < file_duration <= 15 * k:
        print('Film nagrywany pojedynczo')
        idx_start, idx_end = cut_from_timestamps(start_time, start_time + duration, timestamps)

        # wycięcie
        clip = events[idx_start:idx_end]
        # zapisanie
        np.save(destination + '/' + f'{filename[:-7]}_{0}.npy', clip)
        print(f'Clip from {filename} number {0} is done!')
        return None

    if 15 * k < file_duration < 60 * k :
        print(f'Za krótki film:{file_duration // k} s')
        print('Brak podziału na części. Rozpatrywanie jako całość')

        n = int(((file_duration / k) - 10 ) / 5) # ilość klipów która zmieści się w całym filmie

        for t in range(n):
            clip_start = start_time + t * duration # początek t klipu
            clip_end = clip_start + duration # koniec t klipu

            idx_start, idx_end = cut_from_timestamps(clip_start, clip_end, timestamps=timestamps, idx_start = idx_start, idx_end=idx_end)

            # wycięcie
            clip = events[idx_start:idx_end]
            # zapisanie
            np.save(destination + '/' + f'{filename[:-7]}_{t}.npy', clip)
            print(f'Clip from {filename} number {t} is done!')
        return None

    if 60 * k <= file_duration <= 90 * k:
        # Filmik nagrany w serii 10 klipów z przerwą w środku
        # 10 części z przerwą w środku, stąd 2 party
        for t in range(10):
            # part 1
            if t < 5:
                clip_start = start_time + t * duration # początek t klipu
                clip_end = clip_start + duration # koniec t klipu

                idx_start, idx_end = cut_from_timestamps(clip_start, clip_end, timestamps=timestamps, idx_start = idx_start, idx_end=idx_end)

                # wycięcie
                clip = events[idx_start:idx_end]
                # zapisanie
                np.save(destination + '/' + f'{filename[:-7]}_{t}.npy', clip)
                print(f'Clip from {filename} number {t} is done!')

            #part 2
            if t >= 5:
                clip_start_2 = start_2_time + ((t % 5) * duration) # początek t klipu drugiej części
                clip_end_2 = clip_start_2 + duration # koniec t klipu drugiej części

                idx_start, idx_end = cut_from_timestamps(clip_start_2, clip_end_2, timestamps = timestamps, idx_start=idx_start, idx_end=idx_end)

                # wycięcie
                clip = events[idx_start:idx_end]
                # zapisanie
                np.save(destination + '/' + f'{filename[:-7]}_{t}.npy', clip)
                print(f'Clip from {filename} number {t} is done!')

    # Filmik bez przerwy w środku, czynność ciągła w podwójnej serii
    if file_duration > 90 * k:
        print('Plik nagrany w podwójnej serii bez przerw')
        print('Brak podziału na części. Rozpatrywanie jako całość')

        for t in range(20):
        # part 1
            clip_start = start_time + t * duration # początek t klipu
            clip_end = clip_start + duration # koniec t klipu

            idx_start, idx_end = cut_from_timestamps(clip_start, clip_end, timestamps=timestamps, idx_start = idx_start, idx_end=idx_end)

            # wycięcie
            clip = events[idx_start:idx_end]
            # zapisanie
            np.save(destination + '/' + f'{filename[:-7]}_{t}.npy', clip)
            print(f'Clip from {filename} number {t} is done!')
        return None
```

```python
# REGION OF INTERESTS
def _roi_bounding_box(self, roi_mask):
        x, y, width, height = cv.boundingRect(roi_mask.astype(np.uint8))
        return x, y, width, height

def _generate_roi(self, blobized_image, image, ratio_thr=0.5):
    count, obj = cv.connectedComponents(blobized_image, connectivity=4)
    combined_mask = None
    for i in range(1, count):
        obj_mask = obj == i
        x, y, width, height = self._roi_bounding_box(obj_mask)
        # background = image.mean()
        # extracted = (image * obj_mask)[y : y + height, x : x + width] - background
        if (
            ((obj_mask.sum() / (width * height)) > ratio_thr)
            # and ((extracted.sum() / (width * height)) > (0.35 * ratio_thr))
            and (width >= (blobized_image.shape[1] * 0.1))
            and (height >= (blobized_image.shape[0] * 0.1))
        ):
            if combined_mask is None:
                combined_mask = obj_mask
            else:
                combined_mask += obj_mask
    return combined_mask
```
